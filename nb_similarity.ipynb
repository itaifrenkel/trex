# Tree Ensemble Feature Representations
This notebook computes the similarity between pairs of training instances under different train instance feature representations and
similarity metrics.

```python
import argparse
import numpy as np
import matplotlib.pyplot as plt
import lightgbm as lgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris, load_breast_cancer, load_wine
from sklearn.metrics.pairwise import linear_kernel, rbf_kernel, euclidean_distances
from sexee.explainer import TreeExtractor
```

```python
def similarities(model='lgb', dataset='iris', n_estimators=20, random_state=69, color='green', fname=None):

    # create model
    if model == 'lgb':
        clf = lgb.LGBMClassifier(random_state=random_state, n_estimators=n_estimators)
    elif model == 'rf':
        clf = RandomForestClassifier(random_state=random_state, n_estimators=n_estimators)

    # load dataset
    if dataset == 'iris':
        data = load_iris()
    elif dataset == 'breast':
        data = load_breast_cancer()
    elif dataset == 'wine':
        data = load_wine()

    X = data['data']
    y = data['target']
    label = data['target_names']

    # train a tree ensemble
    clf = clf.fit(X, y)
    extractor_path = TreeExtractor(clf, encoding='tree_path')
    extractor_output = TreeExtractor(clf, encoding='tree_output')

    X_dist_sim = euclidean_distances(X)

    X_path = extractor_path.fit_transform(X)
    X_path_sim = linear_kernel(X_path)

    X_out = extractor_output.fit_transform(X)
    X_out_sim = rbf_kernel(X_out)
    
    # compute correlations
    dist_path_corr = np.corrcoef(X_dist_sim.flatten(), X_path_sim.flatten())[0][1]
    dist_out_corr = np.corrcoef(X_dist_sim.flatten(), X_out_sim.flatten())[0][1]
    path_out_corr = np.corrcoef(X_path_sim.flatten(), X_out_sim.flatten())[0][1]

    # plot correlation between feature representation similarities
    fig, axs = plt.subplots(1, 3, figsize=(20, 5))
    fig.suptitle('Feature Representation Similarity, Model: {}, Dataset: {}'.format(model, dataset))
    axs[0].scatter(X_dist_sim.flatten(), X_path_sim.flatten(), color=color)
    axs[0].set_title('Original vs Tree Path (corr: {:.3f})'.format(dist_path_corr))
    axs[0].set_xlabel('original (euclidean)')
    axs[0].set_ylabel('tree path (linear kernel)')
    axs[1].scatter(X_dist_sim.flatten(), X_out_sim.flatten(), color=color)
    axs[1].set_title('Original vs Tree Output (corr: {:.3f})'.format(dist_out_corr))
    axs[1].set_xlabel('original (euclidean)')
    axs[1].set_ylabel('tree output (rbf kernel)')
    axs[2].scatter(X_path_sim.flatten(), X_out_sim.flatten(), color=color)
    axs[2].set_title('Tree Path vs Tree Output (corr: {:.3f})'.format(path_out_corr))
    axs[2].set_xlabel('tree path (linear kernel)')
    axs[2].set_ylabel('tree output (rbf kernel)')
    
    if fname is not None:
        plt.savefig(fname, format='pdf', bbox_inches='tight')
    else:
        plt.show()
```

## LGB Model

```python
similarities(model='lgb', dataset='iris', color='green')
similarities(model='lgb', dataset='breast', color='orange')
similarities(model='lgb', dataset='wine', color='red')
```

## RF Model

```python
similarities(model='rf', dataset='iris', color='green')
similarities(model='rf', dataset='breast', color='orange')
similarities(model='rf', dataset='wine', color='red')
```

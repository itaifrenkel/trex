```python
import json
import catboost
import numpy as np
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data['data']
y = data['target']

# X = np.random.randn(10, 2)
# y = np.random.randint(2, size=10)

cb = catboost.CatBoostClassifier(random_state=69, n_estimators=100, verbose=0).fit(X, y)
```

```python
cb.save_model('modelx.json', format='json')
jx = json.load(open('modelx.json', 'r'))
jx['oblivious_trees'][0]
```

```python
p = catboost.Pool(X)
cb.plot_tree(0, p)
```
